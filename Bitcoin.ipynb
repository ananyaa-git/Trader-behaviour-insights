## Data cleaning

### Subtask:
Clean the trading data by handling outliers and the unusual negative fee values.


**Reasoning**:
Investigate negative fee values and handle outliers in specified columns.  This involves analyzing the 'Fee' column for negative values, identifying and treating outliers in 'Execution Price', 'Size Tokens', 'Size USD', and 'Closed PnL', and documenting the changes.



```python
# Investigate Negative Fees
negative_fees = df[df['Fee'] < 0]
print(f"Number of negative fees: {len(negative_fees)}")
display(negative_fees.head())
# Since the number of negative fees is relatively small and they might represent specific transaction types or errors, replacing them with 0 is a reasonable approach.
df['Fee'] = df['Fee'].clip(lower=0)

# Outlier Handling
import matplotlib.pyplot as plt
cols_to_check = ['Execution Price', 'Size Tokens', 'Size USD', 'Closed PnL']

# Function to detect and handle outliers using the IQR method
def handle_outliers_iqr(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    data[column] = data[column].clip(lower=lower_bound, upper=upper_bound)
    return data

for col in cols_to_check:
    plt.figure(figsize=(8, 6))
    plt.title(f"Distribution of {col} Before Outlier Handling")
    plt.hist(df[col])
    plt.show()
    df = handle_outliers_iqr(df, col)
    plt.figure(figsize=(8, 6))
    plt.title(f"Distribution of {col} After Outlier Handling")
    plt.hist(df[col])
    plt.show()

# Data Validation
print("Descriptive Statistics Before and After Outlier Handling:")
for col in cols_to_check:
    print(f"Column: {col}")
    print("Before:")
    print(df[col].describe())
    print("After:")
    print(df[col].describe())

# Document Changes
print("Data Cleaning Summary:")
print("1. Negative Fees:")
print("   - Replaced negative values in the 'Fee' column with 0.")
print("2. Outliers:")
print("   - Used the IQR method to cap outliers in the following columns:", cols_to_check)
print("   - Visualized the distributions before and after handling.")
```

## Data wrangling

### Subtask:
Create new features from existing ones to capture trends and interactions.


**Reasoning**:
Calculate rolling averages, create interaction features, convert the timestamp column to datetime objects, extract date features, and create lagged features.



```python
import pandas as pd

# Calculate rolling averages
for col in ['Closed PnL', 'Size Tokens', 'Size USD']:
    df[f'{col}_rolling_mean_7'] = df[col].rolling(window=7, min_periods=1).mean()
    df[f'{col}_rolling_mean_30'] = df[col].rolling(window=30, min_periods=1).mean()

# Create interaction features
for col in ['Closed PnL']:
    for window in [7, 30]:
        rolling_col = f'{col}_rolling_mean_{window}'
        df[f'{col}_diff_{rolling_col}'] = df[col] - df[rolling_col]
        df[f'{col}_ratio_{rolling_col}'] = df[col] / df[rolling_col]
        df[f'{col}_product_{rolling_col}'] = df[col] * df[rolling_col]

# Convert 'Timestamp IST' to datetime and extract features
df['Timestamp IST'] = pd.to_datetime(df['Timestamp IST'], format='%d-%m-%Y %H:%M', errors='coerce')
df['day_of_week'] = df['Timestamp IST'].dt.dayofweek
df['hour_of_day'] = df['Timestamp IST'].dt.hour
df['month'] = df['Timestamp IST'].dt.month


# Create lagged features
for col in ['Closed PnL', 'Size Tokens', 'Size USD']:
    for lag in [1, 2, 3]:
        df[f'{col}_lagged_{lag}'] = df[col].shift(lag)

display(df.head())
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Select relevant columns for correlation analysis
cols_to_analyze = ['Closed PnL', 'Closed PnL_rolling_mean_7', 'Closed PnL_rolling_mean_30',
                   'Closed PnL_diff_Closed PnL_rolling_mean_7', 'Closed PnL_ratio_Closed PnL_rolling_mean_7',
                   'Closed PnL_product_Closed PnL_rolling_mean_7', 'Closed PnL_diff_Closed PnL_rolling_mean_30',
                   'Closed PnL_ratio_Closed PnL_rolling_mean_30', 'Closed PnL_product_Closed PnL_rolling_mean_30',
                   'Closed PnL_lagged_1', 'Closed PnL_lagged_2', 'Closed PnL_lagged_3',
                   'Size Tokens', 'Size USD', 'Fee']
df_analysis = df[cols_to_analyze].dropna()

# Correlation Analysis
correlation_matrix = df_analysis.corr()

# Visualize correlations using a heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Trader Performance and Engineered Features')
plt.show()

# Identify highly correlated features
highly_correlated_features = correlation_matrix[abs(correlation_matrix['Closed PnL']) > 0.5]
print("Highly correlated features with Closed PnL:\n", highly_correlated_features['Closed PnL'])

# Statistical Tests (t-test example)
# Example: comparing 'Closed PnL' in two groups (e.g., high vs low 'Size Tokens')
# Replace 'Size Tokens' with an appropriate feature for the split
median_size_tokens = df_analysis['Size Tokens'].median()
group1 = df_analysis[df_analysis['Size Tokens'] > median_size_tokens]['Closed PnL']
group2 = df_analysis[df_analysis['Size Tokens'] <= median_size_tokens]['Closed PnL']

t_statistic, p_value = stats.ttest_ind(group1, group2)
print(f"\nT-test results for Closed PnL vs. Size Tokens:\n t-statistic={t_statistic:.4f}, p-value={p_value:.4f}")
if p_value < 0.05:
    print("There is a statistically significant difference in 'Closed PnL' between the two groups.")
else:
    print("There is no statistically significant difference in 'Closed PnL' between the two groups.")

# Time-Series Analysis (rolling correlations)
rolling_correlation = df_analysis['Closed PnL'].rolling(window=30, min_periods=1).corr(df_analysis['Closed PnL_lagged_1'])
plt.figure(figsize=(10,6))
plt.plot(rolling_correlation)
plt.title('30-Day Rolling Correlation between Closed PnL and its 1-day Lag')
plt.xlabel('Time')
plt.ylabel('Correlation')
plt.show()
```

## Feature engineering

### Subtask:
Feature Engineering for Improved Model Performance


**Reasoning**:
Create cyclical features for the hour of day and day of the week using sine and cosine transformations. Create interaction features between the new time features and trader performance metrics. Prepare a list of all engineered features.



```python
import numpy as np

# 1. Time-Based Features
df['sin_hour'] = np.sin(2 * np.pi * df['hour_of_day'] / 24)
df['cos_hour'] = np.cos(2 * np.pi * df['hour_of_day'] / 24)
df['sin_day'] = np.sin(2 * np.pi * df['day_of_week'] / 7)
df['cos_day'] = np.cos(2 * np.pi * df['day_of_week'] / 7)

# 2. Lagged Features Enhancement (already done in previous step)

# 3. Interaction Feature Refinement
performance_metrics = ['Closed PnL', 'Size Tokens', 'Size USD', 'Fee']
time_features = ['sin_hour', 'cos_hour', 'sin_day', 'cos_day']
for metric in performance_metrics:
    for time_feat in time_features:
        df[f'{metric}_{time_feat}_interaction'] = df[metric] * df[time_feat]

# 4. Feature Selection Preparation
engineered_features = [
    'sin_hour', 'cos_hour', 'sin_day', 'cos_day',
    'Closed PnL_rolling_mean_7', 'Closed PnL_rolling_mean_30',
    'Size Tokens_rolling_mean_7', 'Size Tokens_rolling_mean_30',
    'Size USD_rolling_mean_7', 'Size USD_rolling_mean_30',
    'Closed PnL_diff_Closed PnL_rolling_mean_7',
    'Closed PnL_ratio_Closed PnL_rolling_mean_7',
    'Closed PnL_product_Closed PnL_rolling_mean_7',
    'Closed PnL_diff_Closed PnL_rolling_mean_30',
    'Closed PnL_ratio_Closed PnL_rolling_mean_30',
    'Closed PnL_product_Closed PnL_rolling_mean_30',
    'Closed PnL_lagged_1', 'Closed PnL_lagged_2', 'Closed PnL_lagged_3',
    'Size Tokens_lagged_1', 'Size Tokens_lagged_2', 'Size Tokens_lagged_3',
    'Size USD_lagged_1', 'Size USD_lagged_2', 'Size USD_lagged_3',
    'day_of_week', 'hour_of_day', 'month'
]
engineered_features.extend([f'{metric}_{time_feat}_interaction' for metric in performance_metrics for time_feat in time_features])


# Print or document the engineered features
print("Engineered Features:")
for feature in engineered_features:
    print(feature)
display(df.head())
```
